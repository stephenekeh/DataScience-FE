### Feature Engineering with Project with Pakwheels Dataset

```python
## Loading of Data
```

```python
import pandas as pd
path =  'wheels.csv'
df = pd.read_csv(path)
```

```python
df.info()
```

```python
df.head()
```

```python
df['Company Name'].value_counts()
```

```python
## Creation of New Features
```

```python
df_encoded = df.copy()
df_encoded['Age_vs_Mileage'] = df_encoded['Model Year'] / (df_encoded['Mileage'] + 1)
df_encoded['Engine_Efficiency'] = df_encoded['Engine Capacity'] / (df_encoded['Mileage'] + 1) 
df_encoded['Model_Year_and_Engine_Capacity'] = df_encoded['Model Year'] + df_encoded['Engine Capacity']
df_encoded[['Age_vs_Mileage', 'Engine_Efficiency', 'Model_Year_and_Engine_Capacity']].head()
```

```python
df_encoded.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
df_encoded.head()
```

```python
## Label Encoding of Categorcal Featue
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
categorical_columns = ['Company Name','Model Name','Location','Engine Type','Color', 'Assembly','Body Type', 'Transmission Type','Registration Status']
for column in categorical_columns:
    df_encoded[column] = label_encoder.fit_transform(df[column])

df_encoded.head()
```

```python
df_encoded['Company Name'].unique
```

```python
## Feature Scaling 
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
columns_to_scale = ['Price', 'Model Year', 'Mileage', 'Engine Capacity']
df_encoded[columns_to_scale] = scaler.fit_transform(df_encoded[columns_to_scale])
df_encoded.head()
```

```python
## Feature Selection using Random Forest Algorithm
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(df_encoded.drop('Price', axis=1), df_encoded['Price'])
feature_importances = rf.feature_importances_
feature_importance_df = pd.DataFrame({'Feature': df_encoded.drop('Price', axis=1).columns,
                                      'Importance': feature_importances}).sort_values(by='Importance', ascending=False)

feature_importance_df
```

### Observation 
The output of the feature selection above performed using Random Forest Algorithm,
shows that Model_Year_and_Engine_Capacity is the most important feature, followed by Model Year,
Company Name, Model Name and Engine Capacity.
